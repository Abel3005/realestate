```python
import pandas as pd
apart_allRentView = pd.read_sql_query("select * from apart_allRentView", con=engine)
notapart_allRentView = pd.read_sql_query("select * from notapart_allRentView", con=engine)
ansimView = pd.read_sql_query("select * from ansimView", con=engine)
allLhView = pd.read_sql_query("select * from allLhView", con=engine)
# result = pd.read_sql_query(sql_query, con=engine)
```

```python
maeavg=allLhView[allLhView['kindRental']=='매입임대'].groupby(['adresGu'],as_index=False)[['alldeposit']].mean()
hangavg=allLhView[allLhView['kindRental']=='행복주택'].groupby(['adresGu'],as_index=False)[['alldeposit']].mean()
apart_allRentView.rename(columns={'자치구':'adresGu', '전세가':'alldeposit'},inplace=True)
notapart_allRentView.rename(columns={'자치구':'adresGu', '전세가':'alldeposit'},inplace=True)
```

```python
maeavg.alldeposit=maeavg.alldeposit.astype(int)
hangavg.alldeposit=hangavg.alldeposit.astype(int)
apart_allRentView.alldeposit=apart_allRentView.alldeposit.astype(int)
notapart_allRentView.alldeposit=notapart_allRentView.alldeposit.astype(int)
```

```python
# pd.merge([maeavg,notapart_allRentView])
vs_mae_not=pd.merge(maeavg,notapart_allRentView, on ='adresGu', how ='left')
vs_mae_not.rename(columns={'adresGu':'자치구','alldeposit_x':'매입임대 평균가', 'alldeposit_y':'비아파트 평균가'},inplace=True)
vs_hang_apt=pd.merge(hangavg,apart_allRentView, on ='adresGu', how ='left')
vs_hang_apt.rename(columns={'adresGu':'자치구','alldeposit_x':'행복주택 평균가', 'alldeposit_y':'아파트 평균가'},inplace=True)
```

```python
# 한글 폰트 경로
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"

# 폰트 설정
font_prop = fm.FontProperties(fname=font_path)
font_name = font_prop.get_name()
rc('font', family=font_name) # matplotlib으로 구현 될때 한글이 깨지지 않게 해줌

font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf') # 시스템에 설치된 모든 TrueType 폰트 파일의 경로가 리스트 형태로 반환
pd.set_option('display.max_columns', None) # 데이터프레임의 크기 제한 없애는 함수
```

```python
plt.figure(figsize=(30,10))

# 막대 그래프를 옆으로 이동하여 겹치지 않게 조정
bar_width = 0.35
x = np.arange(len(vs_hang_apt['자치구']))

plt.bar(x - bar_width/3, vs_hang_apt['행복주택 평균가'], color='orange', alpha=0.5, label='행복주택 평균가')
plt.bar(x + bar_width/3, vs_hang_apt['아파트 평균가'], color='blue', alpha=0.5, label='아파트 평균가')

plt.xlabel('자치구', fontsize=20, fontproperties=font_prop)
plt.ylabel('평균가', fontsize=30, fontproperties=font_prop)
plt.title('행복주택과 아파트 평균가 비교', fontsize=50, fontproperties=font_prop)
plt.xticks(x, vs_hang_apt['자치구'], rotation=45, fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)
plt.legend(prop=font_prop)

plt.show()
```
