```python
import re

def korean_to_number(string):
    # 한글 숫자를 숫자로 변환하는 함수
    def korean_to_int(korean):
        unit_map = {'억': 100000000, '천': 10000000, '백': 1000000, '만': 10000}
        korean_number_map = {'영': 0, '일': 1, '이': 2, '삼': 3, '사': 4, '오': 5, '육': 6, '칠': 7, '팔': 8, '구': 9}
        result = 0
        current_number = 0
        for char in korean:
            if char in korean_number_map:
                current_number = korean_number_map[char]
            elif char in unit_map:
                result += current_number * unit_map[char]
                current_number = 0
        result += current_number
        return result

    # 입력된 문자열을 숫자로 변환하는 함수
    def parse_number(string):
        number = 0
        current_string = ""
        for char in string:
            if char.isdigit():
                current_string += char
            else:
                number += korean_to_int(current_string)
                current_string = ""
        number += korean_to_int(current_string)
        return number

    # 정규 표현식을 사용하여 숫자를 추출하고 변환
    match = re.findall(r"\d+|[\u2E80-\u2FD5\u3190-\u319f\u3400-\u4DBF\u4E00-\u9FFF]", string)
    return parse_number(''.join(match))

# 테스트
korean_number = '1억1천7백60만'
number = korean_to_number(korean_number)
print(number)  # 출력: 11760000
```

```python
def hangul_to_num(input_str):
    result = 0
    tmp_result = 0
    num = 0
    NUMBER = "0123456789"  # 숫자에 대한 문자열로 변경
    UNIT = "억천만"  # '억' 추가
    UNIT_NUM = [100000000, 1000, 10000, 10000]  # '억'에 해당하는 값 추가

    tokens = []
    token = ''
    for char in input_str:
        if char in NUMBER:  # 숫자인 경우
            token += char
        else:
            tokens.append(token)
            tokens.append(char)
            token = ''
    if token:
        tokens.append(token)

    for token in tokens:
        check = NUMBER.find(token)  # 숫자일 경우 해당 숫자의 인덱스를 얻음
        if check != -1:  # 숫자인 경우
            num = num * 10 + check  # 숫자를 계속 더함
        else:  # 단위인 경우
            if token in UNIT:  # 단위가 '억', '천만', '백만', '만'인 경우
                tmp_result += (num if num != 0 else 1) * UNIT_NUM[UNIT.find(token)]  # 단위에 해당하는 값을 곱함
                num = 0
    return result + tmp_result + num

# 테스트
input_str = "2억7천0만"
print(input_str)
print(hangul_to_num(input_str))
```
